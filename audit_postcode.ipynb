```python
import xml.etree.cElementTree as ET
from collections import defaultdict
import re
import pprint
OSM_FILE = "./Data/portland_oregon.osm/portland_oregon.osm"
postcode_type_re = re.compile(r'\b\S+\.?$', re.IGNORECASE)

expected_postcode = ["97201","97202","97203","97204","97205","97206","97207","97208","97209","97210",
                    "97211","97212","97213","97214","97215","97216","97217","97218","97219","97220",
                    "97221","97222","97223","97224","97225","97227","97228","97229","97230","97231",
                    "97232","97233","97236","97238","97239","97240","97242","97250","97251","97252",
                    "97253","97254","97256","97258","97266","97267","97268","97269","97280","97281",
                    "97282","97283","97286","97290","97291","97292","97293","97294","97296","97298"]

def audit_postcode_type(postcode_types,postcode_number):
    m = postcode_type_re.search(postcode_number)
    if m:
        postcode_type = m.group()
        if postcode_type not in expected_postcode:
            postcode_types[postcode_type].add(postcode_number)
            
def is_postcode_number(elem):
    return (elem.attrib['k'] == "addr:postcode")

def audit_postcode(osmfile):
    osm_file = open(osmfile,"r",encoding = "utf8")
    postcode_types = defaultdict(set)
    
    for event, elem in ET.iterparse(osm_file,events=("start",)):
        if elem.tag == "node" or elem.tag == "way":
            for tag in elem.iter("tag"):
                if is_postcode_number(tag):
                    audit_postcode_type(postcode_types, tag.attrib['v'])
    osm_file.close()
    pprint.pprint(dict(postcode_types))
    return postcode_types

audit_postcode(OSM_FILE)
```
